```python
# data analysis:- understanding data
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sb
pd.pandas.set_option('display.max_columns',None)
```

```python
dataset=pd.read_csv('train.csv')
dataset.shape
```

```python
dataset.head()
```

```python
# find missing val,
# all numerical val,
#  distribution of numerical value,
#   categorical val,
#   cardinality of categorical variable,
# outlier, relationship b/w independent and dependent val
```

```python
features_with_na=[features for features in dataset.columns if dataset[features].isnull().sum()>0]
features_with_na
```

```python
for feature in features_with_na:
  print(feature,np.round(dataset[feature].isnull().mean(),4),'% missing val')
```

```python
# Now we find if the missing value is realted to out outcome
```

```python
for feature in features_with_na:
  data=dataset.copy()
  # 1=null val 0=not null
  data[feature]=np.where(data[feature].isnull(),1,0)
  # cal mean sales price where information is missing
  data.groupby(feature)['SalePrice'].median().plot.bar()
  plt.title(feature)
  plt.show()
```

With relation b/w missing val and dependent val is visible. So we will replace nan values in feature selection

```python
disc_features=[feature for feature in num_features if len(dataset[feature].unique())<25 and feature not in year_feature+['Id'] ]
cont_features=[feature for feature in num_features if feature not in disc_features and feature not in year_feature+['Id'] ]
disc_features
```

```python
cont_features
```

**for discrete feature we are using bar chart**

```python
for feature in disc_features:
  data=dataset.copy()
  data.groupby(feature)['SalePrice'].median().plot.bar()
  plt.xlabel(feature)
  plt.ylabel('SalePrice')
  plt.title(feature)
  plt.show()
```

**for continous feature we are using hostogram**

```python
for feature in cont_features:
  data=dataset.copy()
  data[feature].hist(bins=25)
  plt.xlabel(feature)
  plt.ylabel('SalePrice')
  plt.title(feature)
  plt.show()
```

**Using Logratithmic Transformation**

```python
for feature in cont_features:
  data=dataset.copy()
  if 0 in data[feature].unique():
    pass
  else:
    data[feature]=np.log(data[feature])
    data['SalePrice']=np.log(data['SalePrice'])
    plt.scatter(data[feature],data['SalePrice'])
    plt.xlabel(feature)
    plt.ylabel('SalePrice')
    plt.title(feature)
    plt.show()
```

**checking outliers**

```python
for feature in cont_features:
  data=dataset.copy()
  if 0 in data[feature].unique():
    pass
  else:
    data[feature]=np.log(data[feature])
    data.boxplot(column=feature)
    # plt.xlabel(feature)
    plt.ylabel(feature)
    plt.title(feature)
    plt.show()
```
